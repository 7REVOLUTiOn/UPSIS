## 4. Пояснительная записка

### Обоснование структуры календарного плана

При разработке календарного плана для проекта **«Веб-сервис конвертации 2D-изображений в 3D-модели»** я применил декомпозицию работ, соответствующую структуре дипломного проекта. Проект разделён на пять основных этапов, отражающих логику исследования: от теоретического анализа до практической реализации и верификации.

**Этап 1 ** заложен в начале, так как выбор моделей глубины и понимание специфики STL критически влияют на всю последующую разработку. Параллельное изучение метрик качества и вычислительной сложности позволяет сразу сформировать критерии выбора оптимальной модели.

**Этап 2 ** включает не только стандартное проектирование БД и UML, но и специфичные задачи: проектирование пайплайна конвертации и формализацию требований к 3D-печати. Это обусловлено тем, что пригодность моделей к печати — ключевое требование проекта.

**Этап 3 ** является наиболее длительным и включает параллельные ветки: backend (модели глубины, генерация STL) и frontend (пользовательский интерфейс). Выделение отдельных задач на постобработку (3.6) и проверку требований печати (3.7) обеспечивает качество конечных 3D-моделей.

**Этап 4 ** удлинен по сравнению со стандартными проектами из-за необходимости физической верификации — печати тестовых моделей и проверки на дефекты. Это уникальная особенность данного проекта, требующая времени на работу с 3D-принтером.

**Этап 5 ** включает подготовку всех материалов для защиты, включая демо-версию системы.

### Методология анализа рисков и принципы определения контрольных точек

При анализе рисков я применил матрицу вероятности/воздействия, адаптированную под специфику ML-проектов с физической верификацией. Большинство идентифицированных рисков связаны с качеством генерации 3D-моделей (R1, R2, R4), что соответствует предметной области дипломного проекта.

Контрольные точки определены в соответствии с методологией вех проекта (milestones), где каждая веха представляет завершение ключевого этапа с конкретными измеримыми результатами. Критерии достижения сформулированы по принципу SMART.

Особое внимание уделено третьей контрольной точке (рабочий прототип), так как она демонстрирует работоспособность основного функционала. Четвёртая веха включает физическую печать моделей — это критерий, отличающий данный проект от чисто программных решений.
