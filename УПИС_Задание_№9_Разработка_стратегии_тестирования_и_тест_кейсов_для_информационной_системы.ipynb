# **Задание №9. Разработка стратегии тестирования и тест-кейсов для информационной системы**

СТРАТЕГИЯ ТЕСТИРОВАНИЯ И ТЕСТ-КЕЙСЫ
Веб-сервис монокулярной оценки глубины и генерации 3D-моделей для 3D-печати

1. СТРАТЕГИЯ ТЕСТИРОВАНИЯ
1.1. Цели и задачи
Целью тестирования является обеспечение соответствия веб-сервиса для монокулярной оценки глубины и генерации STL-моделей функциональным и нефункциональным требованиям, определенным в спецификации требований (SRS), а также выявление критических дефектов до демонстрации системы на защите дипломного проекта.

Основные задачи:
*   Проверка корректности работы пайплайна: Image → Depth Map → Mesh → STL
*   Верификация качества генерации карт глубины моделями MiDaS, DepthAnything, NeRF, Unet
*   Тестирование метрик качества (SSIM, L1-loss, smoothness) и вычислительной сложности (Params, FLOPs)
*   Проверка корректности экспорта в формат STL и валидации на соответствие требованиям 3D-печати
*   Оценка производительности системы при обработке изображений различного разрешения
*   Проверка безопасности системы от несанкционированного доступа и вредоносных файлов
*   Оценка удобства использования интерфейса для целевой аудитории (студенты, исследователи)

1.2. Область тестирования
Включено в тестирование:
*   Функциональность загрузки и валидации изображений (JPG, PNG, до 10 МБ)
*   Выбор и переключение моделей оценки глубины (MiDaS, DepthAnything, NeRF, Unet)
*   Генерация карт глубины и визуализация результатов
*   Преобразование depth map в 3D-mesh и экспорт в STL
*   Постобработка: фильтрация шумов, редукция полигонов, проверка manifold
*   Валидация STL на соответствие требованиям 3D-печати (толщина стенок, замкнутость)
*   Сравнительный анализ моделей с отображением метрик SSIM, L1, FLOPs
*   Производительность системы при обработке изображений 1024x1024
*   Безопасность системы (защита от XSS, вредоносных загрузок, изоляция ML-контейнеров)
*   Кроссбраузерная совместимость и адаптивность интерфейса

Исключено из тестирования:
*   Обучение нейронных сетей (используются предобученные веса)
*   Функциональность сторонних библиотек (PyTorch, OpenCV, trimesh)
*   Физическая инфраструктура GPU-серверов
*   Долгосрочная стабильность системы (тестирование на месяцы/годы)

1.3. Виды тестирования
**Функциональное тестирование**
Метод: Комбинация ручного и автоматизированного тестирования
Охват: Все требования с приоритетом "Критический" и "Высокий"
Ответственный: Студент-разработчик

**Интеграционное тестирование**
Подход: "Снизу вверх" (Bottom-up)
Фокус: Интеграция ML-моделей с backend, работа с файловой системой, взаимодействие Docker-контейнеров
Инструменты: pytest, Postman, Docker Compose

**Тестирование производительности**
Виды: Нагрузочное, стрессовое
Инструменты: Apache JMeter, Locust
Критерии: Время генерации depth map ≤ 10 сек для изображения 1024x1024 на GPU

**Тестирование безопасности**
Виды: Проверка валидации файлов, тестирование на уязвимости (OWASP Top 10)
Инструменты: OWASP ZAP, Bandit (Python security linter)
Ответственный: Студент с консультацией специалиста по безопасности

**Тестирование юзабилити**
Метод: Тестирование с привлечением реальных пользователей (студенты, научный руководитель)
Цель: Оценка удобства интерфейса для целевой аудитории
Критерии: Не менее 90% успешного выполнения основных сценариев за ≤ 3 минуты

**Регрессионное тестирование**
Частота: После каждого исправления дефектов и добавления новой функциональности
Автоматизация: 70% регрессионных тестов автоматизировано
Инструменты: pytest, Selenium

1.5. Критерии входа и выхода
**Entry Criteria (Критерии начала тестирования):**
*   Код зафиксирован в системе контроля версий (Git)
*   Сборка развернута на тестовом стенде (Docker Compose)
*   Тестовая документация (тест-кейсы) подготовлена и утверждена руководителем
*   Тестовые данные созданы и доступны (набор vintage photos, тестовые изображения)
*   Выполнены unit-тесты с покрытием ≥ 70% кодовой базы
*   Пройден smoke test базовой функциональности (загрузка → обработка → скачивание)

**Exit Criteria (Критерии завершения тестирования):**
*   Выполнено ≥ 95% запланированных тест-кейсов
*   Покрытие требований с приоритетом "Критический" = 100%
*   Покрытие требований с приоритетом "Высокий" ≥ 90%
*   Pass Rate ≥ 98% для критических и высокоприоритетных тестов
*   Все дефекты с серьезностью "Critical" исправлены
*   Все дефекты с серьезностью "Major" исправлены или отложены с обоснованием
*   Нагрузочные тесты пройдены успешно (время обработки ≤ 10 сек на изображение)
*   Тесты безопасности не выявили критических уязвимостей
*   Юзабилити-тестирование показало успешное выполнение ≥ 90% сценариев

1.6. Метрики качества
**Метрики покрытия:**
*   Покрытие требований: ≥ 90% для критических и высокоприоритетных требований
*   Покрытие кода unit-тестами: ≥ 70%
*   Покрытие функциональных сценариев: ≥ 85%

**Метрики дефектов:**
*   Плотность дефектов: < 2 дефекта/1000 строк кода
*   Эффективность обнаружения дефектов (DDE): ≥ 90%
*   Распределение дефектов по серьезности:
    *   Critical: 0 (должны быть исправлены)
    *   Major: ≤ 5%
    *   Minor и Trivial: оставшиеся

**Метрики выполнения:**
*   Процент выполненных тестов: ≥ 95%
*   Pass Rate: ≥ 98% для критических тестов
*   Скорость исправления дефектов:
    *   Critical: < 24 часа
    *   Major: < 72 часа
    *   Minor: < 1 неделя
